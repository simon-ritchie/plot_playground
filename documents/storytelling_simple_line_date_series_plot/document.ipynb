# Sample data

```python
from datetime import datetime

import pandas as pd
from pandas.tseries.offsets import Day, MonthBegin
import numpy as np
```

```python
start_dt = datetime(2015, 6, 1)
last_dt = datetime(2019, 2, 1)

date_str_list = []

orange_shipment_list = []
apple_shipment_list = []
melon_shipment_list = []

orange_price_list = []
apple_price_list = []
melon_price_list = []

index = 0
target_dt = start_dt
while target_dt <= last_dt:
    target_date_str = target_dt.strftime('%Y-%m')
    date_str_list.append(target_date_str)
    orange_shipment_list.append(np.random.randint(2500, 3000))
    index_adjustment_val = index * 50
    apple_shipment_list.append(
        int(np.random.randint(1700 + index_adjustment_val, 2000 + index_adjustment_val)))
    melon_shipment_list.append(np.random.randint(1100, 1500))
    
    orange_price_list.append(np.random.randint(130, 180) / 100)
    apple_price_list.append(np.random.randint(120, 170) / 100)
    melon_price_list.append(np.random.randint(250, 450) / 100)
    target_dt += MonthBegin()
    index += 1
```

```python
df = pd.DataFrame(
    columns=['date', 'orange_shipment', 'orange_price', 'apple_shipment',
             'apple_price', 'melon_shipment', 'melon_price'],
    index=np.arange(0, len(date_str_list)))
```

```python
df.date = date_str_list
df.orange_shipment = orange_shipment_list
df.orange_price = orange_price_list
df.apple_shipment = apple_shipment_list
df.apple_price = apple_price_list
df.melon_shipment = melon_shipment_list
df.melon_price = melon_price_list
```

```python
df
```

# Simple use case

```python
from plot_playground.storytelling.simple_line_date_series_plot import display_plot
```

```python
display_plot(
    df=df,
    date_column='date',
    normal_columns=['apple_shipment', 'melon_shipment'], 
    stands_out_columns=['orange_shipment'],
    line_color='#e8e8e8')
```
